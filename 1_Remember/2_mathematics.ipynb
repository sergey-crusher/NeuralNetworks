### Основные правила дифференцирования

##### 1. Константа
$\huge \frac{d}{dx}(c) = 0$

##### 2. Степенная функция
$\huge \frac{d}{dx}(x^n) = n \cdot x^{n-1}$

##### 3. Константа × функция
$\huge \frac{d}{dx}(c \cdot f(x)) = c \cdot f'(x)$

##### 4. Сумма и разность
$\huge \frac{d}{dx}(f(x) \pm g(x)) = f'(x) \pm g'(x)$

##### 5. Произведение
$\huge \frac{d}{dx}(f(x) \cdot g(x)) = f'(x) \cdot g(x) + f(x) \cdot g'(x)$

##### 6. Частное
$\huge \frac{d}{dx}\left(\frac{f(x)}{g(x)}\right) = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{g(x)^2}$

##### 7. Сложная функция (цепное правило)
$\huge \frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$

##### 8. Элементарные функции

| $\huge f(x)$     | $\huge f'(x)$                   |
|------------------|---------------------------------|
| $\huge \sin x$   | $\huge \cos x$                  |
| $\huge \cos x$   | $\huge -\sin x$                 |
| $\huge \tan x$   | $\huge \frac{1}{\cos^2 x}$      |
| $\huge e^x$      | $\huge e^x$                     |
| $\huge a^x$      | $\huge a^x \cdot \ln a$         |
| $\huge \ln x$    | $\huge \frac{1}{x}$             |
| $\huge \log_a x$ | $\huge \frac{1}{x \cdot \ln a}$ |
