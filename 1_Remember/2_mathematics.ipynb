### Основные правила дифференцирования

##### 1. Константа
$\huge \frac{d}{dx}(c) = 0$

##### 2. Степенная функция
$\huge \frac{d}{dx}(x^n) = n \cdot x^{n-1}$

##### 3. Константа × функция
$\huge \frac{d}{dx}(c \cdot f(x)) = c \cdot \frac{d}{dx}f(x)$

##### 4. Сумма и разность
$\huge \frac{d}{dx}(f(x) \pm g(x)) = \frac{d}{dx}f(x) \pm \frac{d}{dx}g(x)$

##### 5. Произведение
$\huge \frac{d}{dx}(f(x) \cdot g(x)) = \frac{d}{dx}f(x) \cdot g(x) + f(x) \cdot \frac{d}{dx}g(x)$

##### 6. Частное
$\huge \frac{d}{dx}\left(\frac{f(x)}{g(x)}\right) = \frac{\frac{d}{dx}f(x) \cdot g(x) - f(x) \cdot \frac{d}{dx}g(x)}{g(x)^2}$

##### 7. Сложная функция (цепное правило)
$\huge \frac{d}{dx}f(g(x)) = \frac{df}{dg} \cdot \frac{dg}{dx}$

Пример с логарифмом

$\huge \frac{d}{dx} \ln(x^2 - 1)$

**Шаг 1.** Внешняя функция: $\ln(u)$, производная $\huge \frac{1}{u}$

**Шаг 2.** Внутренняя функция: $u = x^2 - 1$, производная $2x$

**Шаг 3.** Перемножаем:

$\huge \frac{1}{x^2 - 1} \cdot 2x = \frac{2x}{x^2 - 1}$

##### 8. Элементарные функции

| $\huge f(x)$     | $\huge \frac{d}{dx}f(x)$        |
|------------------|---------------------------------|
| $\huge \sin x$   | $\huge \cos x$                  |
| $\huge \cos x$   | $\huge -\sin x$                 |
| $\huge \tan x$   | $\huge \frac{1}{\cos^2 x}$      |
| $\huge e^x$      | $\huge e^x$                     |
| $\huge a^x$      | $\huge a^x \cdot \ln a$         |
| $\huge \ln x$    | $\huge \frac{1}{x}$             |
| $\huge \log_a x$ | $\huge \frac{1}{x \cdot \ln a}$ |

В следующем примере будет натуральный логарифм, по этому пройдёмся сначала по теории.

Логарифм — это ответ на вопрос: "В какую степень нужно возвести основание, чтобы получить число?".

$\log_2 8 = 3$

Если логарифм записывают как $\ln x$, это натуральный логарифм со степенью $e \approx 2.71828$ (число Эйлера).

##### Откуда взялось $e$ (исторически)
В XVII веке математики изучали сложные проценты.
**Якоб Бернулли** (швейцарский математик) в 1683 году исследовал предел $\huge \left(1 + \frac{1}{n}\right)^n$ в задаче о непрерывном начислении процентов. Он первым обнаружил, что этот предел сходится к числу между 2 и 3.

**Леонард Эйлер** позже (1731 год) обозначил эту константу буквой $e$.

Упростим задачу до простого вопроса: если положить 1 рубль под 100% годовых, сколько будет через год?
##### Простые проценты (1 начисление в конце года)
$1 + 1 = 2 \quad \text{рубля}$
##### Сложные проценты (начисление каждые полгода)
За полгода начисляем 50%:
$1 \cdot \left(1 + \frac{1}{2}\right)^2 = 2.25 \quad \text{рубля}$
##### Начисление каждый месяц
$1 \cdot \left(1 + \frac{1}{12}\right)^{12} \approx 2.613 \quad \text{рубля}$
##### Начисление каждый день
$1 \cdot \left(1 + \frac{1}{365}\right)^{365} \approx 2.714 \quad \text{рубля}$
##### А если проценты начисляются непрерывно?
Математики задались вопросом: к чему стремится выражение, если начисление происходит всё чаще и чаще, мгновенно?

$\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = 2.718281828459... = e$

Этот предел не растёт до бесконечности — он сходится к конкретному числу. Так открыли $e$.
